```python
# COMMAND ----------
from pyspark.sql import functions as F

# ============================================================
# DIM_PRODUCT
# ============================================================
dim_product = spark.table("pharma_catalog.silver.silver_products") \
    .select(
        F.col("product_id"),
        F.col("product_name"),
        F.col("category"),
        F.col("manufacturer"),
        F.col("unit_price"),
        F.col("status")
    )

dim_product.write.mode("overwrite").saveAsTable("pharma_catalog.gold.dim_product")
print(f"✅ dim_product : {dim_product.count()} lignes")

# COMMAND ----------
# ============================================================
# DIM_SUPPLIER
# ============================================================
dim_supplier = spark.table("pharma_catalog.silver.silver_suppliers") \
    .select(
      F.col("supplier_id"),
        F.col("supplier_name"),
        F.col("country"),
        F.col("lead_time_days"),
        F.col("contact_email"),
        F.col("status")
    )

dim_supplier.write.mode("overwrite").saveAsTable("pharma_catalog.gold.dim_supplier")
print(f"✅ dim_supplier : {dim_supplier.count()} lignes")

# COMMAND ----------
# ============================================================
# DIM_WAREHOUSE
# ============================================================
dim_warehouse = spark.table("pharma_catalog.silver.silver_warehouses") \
    .select(
        F.col("warehouse_id"),
        F.col("warehouse_name"),
        F.col("country"),
        F.col("city"),
        F.col("capacity_units"),
        F.col("manager_name")
    )

dim_warehouse.write.mode("overwrite").saveAsTable("pharma_catalog.gold.dim_warehouse")
print(f"✅ dim_warehouse : {dim_warehouse.count()} lignes")

# COMMAND ----------
# ============================================================
# DIM_DATE
# ============================================================
dim_date = spark.table("pharma_catalog.silver.silver_calendar") \
    .select(
        F.col("date_key"),
        F.col("full_date"),
        F.col("day"),
        F.col("month"),
        F.col("month_name"),
        F.col("quarter"),
        F.col("year"),
        F.col("week_number")
    )

dim_date.write.mode("overwrite").saveAsTable("pharma_catalog.gold.dim_date")
print(f"✅ dim_date : {dim_date.count()} lignes")

# COMMAND ----------
# Vérification finale
print("=== GOLD DIMENSIONS ===")
for table in ["dim_product", "dim_supplier", "dim_warehouse", "dim_date"]:
    count = spark.table(f"pharma_catalog.gold.{table}").count()
    print(f"  {table} : {count} lignes")
```
