```python
# COMMAND ----------
# 01_landing_to_bronze.py
# Copie des CSV vers le Landing Volume

import shutil
import os

git_data_path = "/Workspace/Users/glennepoaty@gmail.com/pharma-supply-chain-analytics/Data/raw/"
landing_path = "/Volumes/pharma_catalog/landing/raw/"

files = os.listdir(git_data_path)

for file in files:
    src = git_data_path + file
    dst = landing_path + file
    shutil.copy(src, dst)
    print(f"✅ Copié : {file}")

print("\n Tous les fichiers sont dans le Landing Volume !")

# COMMAND ----------
# Vérification
print(os.listdir(landing_path))
```

```python
# COMMAND ----------
# Configuration
landing_path = "/Volumes/pharma_catalog/landing/raw/"

tables = [
    "products",
    "suppliers",
    "warehouses",
    "inventory",
    "sales_orders",
    "supplier_orders",
    "calendar",
    "alerts"
]

# COMMAND ----------
# Ingestion CSV → Tables Bronze

for table in tables:
    df = spark.read.option("header", "true").option("inferSchema", "true").csv(f"{landing_path}{table}.csv")
    
    df.write.mode("overwrite").saveAsTable(f"pharma_catalog.bronze.bronze_{table}")
    
    print(f"✅ bronze_{table} chargée — {df.count()} lignes")

print("\n🎉 Toutes les tables Bronze sont créées !")

# COMMAND ----------
# Vérification
spark.sql
spark.sql("SHOW TABLES IN pharma_catalog.bronze").show()
```
